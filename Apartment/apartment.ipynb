### Открыл файл с данными и изучил общую информацию. 

```python
import pandas as pd
#библиотека для создания статистических графиков
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./real_estate_data.csv', sep='\t')
data.head(10)
```

У нас 22 столбца с данными:

* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)

* living_area — жилая площадь в квадратных метрах(м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км

* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

### Предобработка данных

```python
data.info()
```

```python
#провекра на пропуски
data.isnull().sum()
```

Заменим пропуски в столбцах высота потолков, жилая площадь, площадь кухни на среднее значение. В столбце Этажи заменим на 0, так как их может и не быть. В столбце Название населенного пункта заменим на Неизвестно, так как дома могут быть расположены не в населонном пункте. В days_exposition на 0, объявление может быть выложенно меньше дня. В cityCenters_nearest на 0, дома могут быть в самом центре города.

```python
#замена пропусков средним значением и 0
celing_median = data['ceiling_height'].median()
living_median = data['living_area'].median()
kitchen_median = data['kitchen_area'].median()
airports_median = data['airports_nearest'].median()
values = {'ceiling_height': celing_median,'floors_total': 0, 
          'balcony': 0, 'living_area': living_median, 'kitchen_area': kitchen_median, 
          'locality_name': 'Неизвестно', 'days_exposition': 0, 'cityCenters_nearest': 0}
data.fillna(value=values, inplace=True)
data.isnull().sum()
```

Теперь поменяем тип данных:

```python
data['floors_total'] = data['floors_total'].astype(int) 
data['is_apartment'] = data['is_apartment'].astype(bool)
data['balcony'] = data['balcony'].astype(int)
data['cityCenters_nearest'] = data['cityCenters_nearest'].astype(int)
data['days_exposition'] = data['days_exposition'].astype(int)
data.info()
```

### Проведем классификацию некоторых столбцов и добавим в таблицу 

```python
#ценна квадратного метра
data['price_metr'] = data['last_price'] / data['total_area']
data.head(10)
```

```python
#перевели в формат datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

#создали новые столбца день недели, месяц и год публикации объявлений
data['day_of_week'] = data['first_day_exposition'].dt.weekday
data['moth_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
data.head(10)
```

```python
#этаж квартиры; варианты — первый, последний, другой
data['floor_kvar'] = 'другой' 
data['floor_kvar'] = data['floor_kvar'].where(data['floor'] != 1, 'первый') 
data['floor_kvar'] = data['floor_kvar'].where(data['floor'] != data['floors_total'], 'последний')
data['floor_kvar'].value_counts()
```

Создали столбец со значениями Другой. Потом с помощью метода where заменили значения 'первый' и 'последний'

```python
#соотношение жилой и общей площади
data['liv_tot'] = data['living_area'] / data['total_area']
#соотношение площади кухни к общей
data['kitch_tot'] = data['kitchen_area'] / data['total_area']
data.head(10)
```

#### Какие факторы больше всего влияют на стоимость квартиры? Изучим, зависит ли цена от площади, числа комнат, удалённости от центра.

Поэтому в середине весны люди выкладывают объявления с максимальной стоимостью, в надежде найти покупателя до начало лета. Потом постепено снижают ценну, пока не найдут покупателя. На графике видно, что с 2014 года стоимость метра снижается до 2016 года, но потом постепенно увеличивается. Скорее всего это связано с тем, что после обвала рубля в 2014 году повышенный спрос на жилье закончился в начале января 2015 года.

Мы выяснили, что на стоимость квадратного метра больше всего влияет количество комнат, этаж квартиры и удаленность от центра. Зависимости от даты размещения почти нет. Так же узнали, что самая высока стоимость жилья в Санкт-Петербурге, а самая маленькая в Выборге.
В цетральной зоне (6км) Санкт-Петербурга самое дорогое жильё. Стоимость квадратного метра постоянно меняется, возможно это мошенники пытаются ускорить продажу чужой квартиры. Поэтому в центре Санкт-Петербурга есть риск мошенничества. Или же это ,наоборот, люди, которые никак не могу продать квартиру, постоянно меняют её стоимость.
С 2014 года стоимость метра снижается до 2016 года, но потом постепенно увеличивается - это было связано с падением рубля, в то время, как в центре Санкт-Петербурга стоимость жилья увеличивается и не спадает. Это говорит о том, что не всё зависит от внутренних факторах, про внешние также забывать не стоит.
